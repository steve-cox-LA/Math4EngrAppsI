# Trigonometry

We have solved problems involving the motion of cars and the falling of objects. The path of motion in each case has been one-dimensional. If we instead want to follow the trajectory of a cannonball as it leaves a cannon or a paratrooper as she jumps from an airplane we need to work in (at least) two dimensions.

To work in two dimensions engineers, over the years, have relied more and more on the concepts of trigonometry. There are just two key notions, **distance** and **angle**. The former is of course a stowaway from our one-dimensional work. Angle is something new, but in order to truly tame it we express it as a ratio of distances.

In the picture below, $\theta$ is the angle, $r$ is the radius, and $s$ is the length of the arc. This picture illustrates the beautiful relationship
$$
   \boxed{\theta = \frac{s}{r}}  \tag{1}
$$
As a ratio of lengths, $\theta$ has NO dimensions. We say that $\theta$ has units of **radians**, as opposed to the perhaps more common unit of degrees.

```python
import numpy as np                  # import all the numerical math methods
import matplotlib.pyplot as plt     # import all the plotting methods
%matplotlib inline

t = np.linspace(0, np.pi/4, 100)   # make 100 evenly spaced points between 0 and pi/4
plt.plot(np.cos(t), np.sin(t))     # plot an arc of the unit circle (in blue, the default color)

theta = np.pi/4

plt.plot([0, 1],[0, 0],'k')                                                 # plot the black line
plt.text(0.5*np.cos(1.1*theta), 0.5*np.sin(1.1*theta),'r',fontsize=20)
plt.plot([0, np.cos(theta)],[0, np.sin(theta)],'r')                         # plot the red ray
plt.text(1.05*np.cos(theta/2), 1.05*np.sin(theta/2),'s',fontsize=20)
plt.text(0.25*np.cos(theta/2), 0.25*np.sin(theta/2),'$\\theta=s/r$',fontsize=20)
plt.text(0,0.65,'Figure 1',fontsize=20)
plt.axis('off')
plt.axis('equal');
```

You may recall that arc length of a full circle (often called **circumference** or perimeter) of radius $r$ is $2\pi r$. Rather than simply stating this formula, we will be able to derive it from simpler facts.

We next associate two important functions with each angle. Both functions, as with the angle itself, are ratios of (signed) lengths. These lengths come from dropping a vertical down from the end of the arc.  In the natural case of the **unit circle** (i.e., $r=1$) this gives us a right triangle with

$$
  \boxed{\cos(\theta) \equiv \hbox{"base"} \hskip 0.25in\hbox{and}\hskip 0.25in  \sin(\theta) \equiv \hbox{"height"}}
$$

We placed **base** and **height** in quotes because we actually keep track of their signs, depending on which quadrant $\theta$ places us in. To see how **cosine** and **sine** change with $\theta$ please exercise the widget below.

```python
# draw the right triangle inside the unit circle with a widget angle

%matplotlib notebook
import numpy as np                  # import all the numerical math methods
import matplotlib.pyplot as plt     # import all the plotting methods
from ipywidgets import *

fig, ax = plt.subplots()

t = np.linspace(0, 2*np.pi, 1000)  # draw the unit circle
ax.plot(np.cos(t), np.sin(t), 'r')
ax.grid('on')

theta = np.pi/8
s = np.sin(theta)
c = np.cos(theta)

triangle, = ax.plot([0, c, c, 0], [0, 0, s, 0], 'b')
tlab = ax.text(c/4, s/8, '$\\theta$')
clab = ax.text(5*c/8, 0.05, '$\cos(\\theta)$')
slab = ax.text(1.05*c, s/2, '$\sin(\\theta)$')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_aspect('equal')
ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
    
def update(theta=widgets.FloatSlider(min=0, max=2*np.pi, step=np.pi/16, value=np.pi/8)):
    c = np.cos(theta)
    s = np.sin(theta)
    triangle.set_data([0, c, c, 0], [0, 0, s, 0])
    tlab.set_position((c/4, s/8))
    clab.set_position((5*c/8, 0.05))
    slab.set_position((1.05*c, s/2))
    fig.canvas.draw_idle()

interact(update);
```

If our radius is not 1 we simple scale our trig functions by $r$, as in the figure below

```python
%matplotlib inline
theta = np.pi/8
plt.plot([0, np.cos(theta)],[0, 0],'k')  
plt.text(0.5*np.cos(1.1*theta), 0.5*np.sin(1.1*theta),'r',fontsize=20)
plt.plot([0, np.cos(theta)],[0, np.sin(theta)],'r')
plt.plot([np.cos(theta), np.cos(theta)],[0, np.sin(theta)],'g')
plt.text(1.05*np.cos(theta), 0.5*np.sin(theta),'$r*\sin(\\theta)$',fontsize=20)
plt.text(0.35*np.cos(theta), -.1,'$r*\cos(\\theta)$',fontsize=20)
plt.text(0.25*np.cos(theta/2), 0.25*np.sin(theta/2),'$\\theta$',fontsize=20)
plt.axis('off')
plt.axis('equal');
```

You may also know this picture in wordier terms, like those below

```python
%matplotlib inline
plt.plot([0, np.cos(theta)],[0, 0],'k') 
plt.text(0.5*np.cos(1.1*theta)-.2, 0.5*np.sin(1.1*theta),'hyp',fontsize=20)
plt.plot([0, np.cos(theta)],[0, np.sin(theta)],'r')
plt.plot([np.cos(theta), np.cos(theta)],[0, np.sin(theta)],'g')
plt.text(1.05*np.cos(theta), 0.5*np.sin(theta),'opp',fontsize=20)
plt.text(0.35*np.cos(theta), -.1,'adj',fontsize=20)
plt.text(0.25*np.cos(theta/2), 0.25*np.sin(theta/2),'$\\theta$',fontsize=20)
plt.axis('off')
plt.axis('equal');
```

where hyp is short for **hypotenuse**, opp for **opposite** and adj for **adjacent**. These are each (signed) lengths and we define our trig functions via
$$
  \cos(\theta) = \frac{\hbox{adj}}{\hbox{hyp}} \quad\hbox{and}\hskip 0.25in \sin(\theta) = \frac{\hbox{opp}}{\hbox{hyp}}
$$
As our angle $\theta$ moves from $0$ to $2\pi$ these functions trace out the beatiful curves below

```python
%matplotlib inline
theta = np.linspace(0, 2*np.pi, 1000)  
plt.plot(theta, np.cos(theta), 'r')
plt.plot(theta, np.sin(theta), 'b')
plt.xlabel('$\\theta$')
plt.legend(['cos','sin'])
plt.grid('on')
```

From this figure we immediately read off the angles at which our trig functions reach their extremes

$$
  \cos(0) = 1, \hskip 0.25in \sin(0) = 0 \\
  \cos(\pi/2) = 0, \hskip 0.25in\sin(\pi/2) = 1 \\
  \cos(\pi) = -1, \hskip 0.25in \sin(\pi) = 0 \\
  \cos(3\pi/2) = 0, \hskip 0.25in \sin(3\pi/2) = -1
$$

Do you **see** that when one trig function attains it's extreme the other trig function vanishes?

I wish to emphasize that these are not things we memorize, but things we **see** and **know**.

So far we have followed the convention of measuring angles as positive as we move counterclockwise from the positive $x$ axis. If we reverse our direction it makes sense to reverse the sign of our angle. If we reflect our second figure over the $x$-axis we find

```python
%matplotlib inline
theta = np.pi/4
plt.plot([0, np.cos(theta)],[0, 0],'k')   
plt.text(0.5*np.cos(-1.1*theta)-0.1, 0.5*np.sin(-1.1*theta),'r',fontsize=20)
plt.plot([0, np.cos(-theta)],[0, np.sin(-theta)],'r')
plt.plot([np.cos(-theta), np.cos(-theta)],[0, np.sin(-theta)],'g')
plt.text(1.05*np.cos(-theta), 0.5*np.sin(-theta),'$r*\sin(-\\theta)$',fontsize=20)
plt.text(0.35*np.cos(-theta), 0.1,'$r*\cos(-\\theta)$',fontsize=20)
plt.text(0.25*np.cos(-theta/2), 0.25*np.sin(-theta/2),'$-\\theta$',fontsize=20)
plt.axis('off')
plt.axis('equal');
```

Comparing the original and reflected triangles we see that the **adjacent leg** remains unchanged, and so
$$
  \boxed{\cos(-\theta) = \cos(\theta)} \tag{3}
$$
while the **opposite leg** is now in the land of negative $y$ and so
$$
  \boxed{\sin(-\theta) = -\sin(\theta)}  \tag{4}
$$
Let us confirm that python knows these facts.

```python
%matplotlib inline
theta = np.linspace(-2*np.pi, 2*np.pi, 1000)
plt.plot(theta, np.cos(theta), 'r')
plt.plot(theta, np.sin(theta), 'b')
plt.legend(['cos','sin'])
plt.grid('on')
```

Now that we can handle negative angles let's next make sure we understand that our words **opposite** and **adjacent** are relative to a chosen angle. Let's first calculate the third angle in our second figure by completing this figure to a rectangle, all of whose angles must be $\pi/2$.

```python
import numpy as np                  # import all the numerical math methods
import matplotlib.pyplot as plt 
%matplotlib inline
theta = np.pi/4
plt.plot([0, np.cos(theta)],[0, 0],'k') 
plt.plot([0, np.cos(theta)],[np.sin(theta), np.sin(theta)],'k--') 
plt.text(0.5*np.cos(1.1*theta), 0.5*np.sin(1.1*theta),'r',fontsize=20)
plt.plot([0, np.cos(theta)],[0, np.sin(theta)],'r')
plt.plot([np.cos(theta), np.cos(theta)],[0, np.sin(theta)],'g')
plt.plot([0,0],[0, np.sin(theta)],'g--')
plt.text(1.05*np.cos(theta), 0.5*np.sin(theta),'$r*\sin(\\theta)$',fontsize=20)
plt.text(0.35*np.cos(theta), -.1,'$r*\cos(\\theta)$',fontsize=20)
plt.text(0.25*np.cos(theta/2), 0.25*np.sin(theta/2),'$\\theta$',fontsize=12)
plt.text(0.25*np.cos(3*theta/2)-0.075, 0.25*np.sin(3*theta/2),'$\\pi/2-\\theta$',fontsize=12)
plt.text(0.85*np.cos(theta)-.075, 0.85*np.sin(theta),'$\\theta$',fontsize=12)
plt.text(0.85*np.cos(theta)-.075, 0.85*np.sin(theta)-.1,'$\\pi/2-\\theta$',fontsize=12)
plt.text(0.6, 0.05,'$\pi/2$', fontsize=12)
plt.axis('off')
plt.axis('equal');
```

Now do you **see** that the opposite to $\theta$ is the adjacent to $\pi/2-\theta$, and vice versa? If so, then you can **see**

$$
  \boxed{\cos(\theta)=\sin(\pi/2-\theta) \quad\hbox{and}\hskip 0.25in \sin(\theta) = \cos(\pi/2-\theta)}  \tag{5}
$$

Finally, let's practice recognizing which of these combinations of reflections and shifts give proper identities.

```python
## pose random true false question between reflected and shifted   cos(theta) and sin(theta)  values
# Press the Run button and work until you can get 20 right in less than a minute

import numpy as np
import time

start_time = time.time()
replies = 0
score = 0
trig = ['cos(theta)', 'sin(theta)', 'cos(-theta)', 'sin(-theta)', 'cos(pi/2 - theta)', 'sin(pi/2-theta)', '-sin(theta)', '-cos(theta)']
ttab = np.eye(8)
ttab[0,[2,5]] = 1
ttab[1,4] = 1
ttab[2,0] = 1
ttab[3,6] = 1
ttab[4,1] = 1
ttab[5,0] = 1
ttab[6,3] = 1
#print(ttab)
while (replies < 20):
    
    m = 2*np.random.randint(4)
    n = 2*np.random.randint(4) + 1
    
    y = ttab[m,n]
    
    try:
       ans = input('True (1) or False (0):  ' + trig[m] + ' = ' + trig[n] + '  ')
       ians = int(ans)
       if ( abs(ians-y) < 0.25 ):
           print('       well done')
           score = score + 1
       else:
           print('       not this time')
       print(' ')
       replies = replies + 1
    except:
        print('oops: I need 1 or 0')
        print(' ')

print('Your Score is ' + str(score) + ' out of 20')
print('You finished in ' + str(round(time.time()-start_time)) + ' seconds')
```
