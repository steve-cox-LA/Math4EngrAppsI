# Quadratic Functions and One-Dimensional Motion

When an object falls from a height $y_0$ its distance to the ground as a function of time is

$$
  y(t) = y_0 - (g/2)t^2, \hskip 0.5in g = 9.8\ m/s^2
$$

You see that it starts at $y(0)=y_0$ and we would like to know how long it takes to hit the ground.

Let's call this time $t_{hit}$ and solve for

$$
  0 = y(t_{hit}) = y_0 - (g/2)t_{hit}^2
$$

we rearrange a bit to arrive at

$$
  (2y_0/g) = t_{hit}^2 \hskip 0.25in \hbox{so}\hskip 0.25in \boxed{t_{hit} = \sqrt{2y_0/g}}
$$

For example, if $y_0=10\, m$ its takes

$$
  t_{hit} = \sqrt{20/9.8} \approx 1.43\ s
$$

to reach the ground.

Its often the case that objects receive some initial oomph, either we toss them up or down with some initial velocity, $v_0$. In this case the distance to the ground satisfies

$$
  y(t) = y_0 + v_0t - (g/2)t^2
$$

and so the time to hit obeys

$$
  y(t_{hit}) = y_0 + v_0t_{hit} - (g/2)t_{hit}^2 = 0
$$

and so, the quadratic formula delivers

$$
  \boxed{t_{hit} = \frac{v_0 + \sqrt{v_0^2 + 2gy_0}}{g}}
$$

Returning to our earlier example, if our object starts at $y_0=10\ m$ and we through it **down** with velocity $v_0=-1\ m/s$
it hits the ground after

$$
  t_{hit} = \frac{-1 + \sqrt{1 + 20*9.8}}{9.8} \approx 1.33\ s
$$

which is sooner than our previous answer, where we simply dropped it with no initial oomph. If instead we tossed it **up** at velocity $v_0=1\ ms$ we arrive at a 

$$
  t_{hit} = \frac{1 + \sqrt{1 + 20*9.8}}{9.8} \approx 1.53\ s
$$

which, of course, is larger than the simple drop time.

Let's now graph the path of our falling objects.

```python
# graph and label a plot of height versus time

import numpy as np                  # numerical helper functions
import matplotlib.pyplot as plt     # graphical helper functions

y0 = 10   # initial height
v0 = 15    # initial velocity
g = 9.8   # gravity
thit = (v0 + np.sqrt(v0**2 + 2*g*y0))/g   # time to hit

t = np.linspace(0,thit,20) 
y = y0 + v0*t - (g/2)*t**2

plt.plot(t, y, 'o-')

plt.grid('on')
plt.xlabel('t (seconds)', fontsize=14)
plt.ylabel('y  (meters)', fontsize=14)
plt.title('Height vs. Time', fontsize=16);
```

In addition to tracking the height of our falling object we can also track its velocity, via

$$
  \boxed{v(t) = v_0 - gt}
$$

You see that height is quadratic in time while velocity is linear in time. This permits us, e.g., to calculate the terminal velocity

$$
  v_{hit} = v(t_{hit}) = v_0 - gt_{hit} = v_0 - g\frac{v_0 + \sqrt{v_0^2 + 2gy_0}}{g} = -\sqrt{v_0^2 + 2gy_0}
$$

So, if $y_0=10$ and $v_0=1$ then the terminal velocity is

$$
  v_{hit} = -\sqrt{1+20g} \approx -14\ m/s
$$

Perhaps you'd like to translate that into mph to best "feel" the impact.

Let's see how to simultaneously graph height and velocity

```python
# graph and label a plot of height and velocity versus time

import numpy as np                  # numerical helper functions
import matplotlib.pyplot as plt     # graphical helper functions

y0 = 10   # initial height
v0 = 5    # initial velocity
g = 9.8   # gravity
thit = (v0 + np.sqrt(v0**2 + 2*g*y0))/g   # time to hit

t = np.linspace(0,thit,20) 
y = y0 + v0*t - (g/2)*t**2

fig, ax1 = plt.subplots()    # prepare for multiple axes
ax2 = ax1.twinx()            # spawn a 2nd set of axes

ax1.plot(t, y, 'bo-')        # plot height on axis 1

v = v0 - g*t
ax2.plot(t, v, 'rs-')        # plot velocity on axis 2

ax1.set_xlabel('t (s)', fontsize=14)
ax1.set_ylabel('height (m)', color='b', fontsize=14)
ax2.set_ylabel('velocity (m/s)', color='r', fontsize=14)
ax1.grid('on')
plt.title('Height & Velocity vs. Time', fontsize=16);
```

Do you see that height reaches its maximum when the velocity is 0? Does this make physical sense to you? 

Let us call the max height $y_{max}$ and the time at which it occurs $t_{max}$. The latter is very easy to calculate.

$$
  0 = v(t_{max}) = v_0 - gt_{max} \hskip 0.25in \hbox{so} \hskip 0.25in \boxed{t_{max} = v_0/g}
$$

and of course this only makes sense if our initiall oomph, $v_0$, is positive. It now follows that

$$
  \boxed{y_{max} = y(t_{max}) = y_0 + v_0t_{max} - (g/2)t_{max}^2 = y_0 + v_0^2/g - v_0^2/(2g) = y_0 + v_0^2/(2g)}
$$

which, given the $y_0=10$ and $v_0=5$ in the graph above yields a

$$
  y_{max} = 10+25/(2*9.8) \approx 11.28\ m \hskip 0.25in \hbox{at} \hskip 0.25in t_{max} = 5/9.8 \approx 0.51\ s
$$

does this indeed agree with our plot?

```python
# one bounce: graph and label a plot of height and velocity versus time

import numpy as np                  # numerical helper functions
import matplotlib.pyplot as plt     # graphical helper functions

y0 = 10   # initial height
v0 = 5    # initial velocity
g = 9.8   # gravity
thit = (v0 + np.sqrt(v0**2 + 2*g*y0))/g   # time to hit

t = np.linspace(0,thit,20) 
y = y0 + v0*t - (g/2)*t**2

fig, ax1 = plt.subplots()    # prepare for multiple axes
ax2 = ax1.twinx()            # spawn a 2nd set of axes

ax1.plot(t, y, 'bo-')        # plot height on axis 1

v = v0 - g*t
ax2.plot(t, v, 'rs-')        # plot velocity on axis 2

# now for the bounce

y0 = 0   # initial height at bounce
v0 = -v[-1]/2    # rebound is opposite of one half of the terminal velocity
g = 9.8   # gravity
thit2 = thit + (v0 + np.sqrt(v0**2 + 2*g*y0))/g   # 2nd time to hit

t = np.linspace(thit,thit2,20)      # new time points
y = y0 + v0*(t-thit) - (g/2)*(t-thit)**2   # new height, with respect to new start time

ax1.plot(t, y, 'bo-')        # plot height on axis 1

v = v0 - g*(t-thit)          # new velocity, with respect to new start time
ax2.plot(t, v, 'rs-')        # plot velocity on axis 2

# label the plots

ax1.set_xlabel('t (s)', fontsize=14)
ax1.set_ylabel('height (m)', color='b', fontsize=14)
ax2.set_ylabel('velocity (m/s)', color='r', fontsize=14)
ax1.grid('on')
plt.title('One Bounce: Height & Velocity vs. Time', fontsize=16);
```
